```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r'C:\Users\Jack-Ryan\Downloads\movies.csv')
```

```python
#lets look at the data

df.head()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
	pct_missing = np.mean(df[col].isnull())
	print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Convert 'budget', 'gross', 'votes', and 'runtime' columns to numeric, replacing non-numeric values with NaN
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['gross'] = pd.to_numeric(df['gross'], errors='coerce')
df['votes'] = pd.to_numeric(df['votes'], errors='coerce')
df['runtime'] = pd.to_numeric(df['runtime'], errors='coerce')


# Convert 'budget', 'gross', 'votes', and 'runtime' columns to int64
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')

# Print data types to verify
print("Data type of 'budget':", df['budget'].dtype)
print("Data type of 'gross':", df['gross'].dtype)
print("Data type of 'votes':", df['votes'].dtype)
print("Data type of 'runtime':", df['runtime'].dtype)
```

```python
df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False )
```
